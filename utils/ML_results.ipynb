```python
### Import Libraries
import numpy as np              # ver_1.26
import pandas as pd             # ver_2.1
from pathlib import Path
from ml_utils import *
```

```python
### Data Path
current_path = Path.cwd()          
project_root = current_path.parents[0]  

data_path = project_root / "data"
image_path = project_root / "result_images"

# print("# data path:", data_path)
```

```python
### input info
feature_cols = ['AGIS','CIGTS','MD_slope_value','MD','VFI_slope_value','VFI', 'TDVmean','TDVstd','PDVmean','PDVstd']
data_cols = feature_cols + ['Exam Age', 'days']
lable_col = ['Consensus_label', 'Wiggs_label']
```

```python
### Load Data
train_df = pd.read_csv(data_path / 'train.csv',sep=',',encoding='cp949')
test_df = pd.read_csv(data_path / 'test.csv',sep=',',encoding='cp949')
```

```python
### Run Code
X_train = train_df[feature_cols]
y_train = train_df[lable_col]
X_test = test_df[feature_cols]
y_test = test_df[lable_col]

models_list = get_models(seed)

df_ty = fit_eval(models_list, X_train, X_test, y_train['Consensus_label'].values, y_test['Consensus_label'].values, "Consensus_label")

df_dy = fit_eval(models_list, X_train, X_test, y_train['Wiggs_label'].values, y_test['Wiggs_label'].values, "Wiggs_label")

results_df = pd.concat([df_ty, df_dy]).set_index(["Target","Model"])

results_df
```

```python
# Save data
results_df.to_csv(data_path / 'RESULT.csv', index=False, encoding='cp949')
```
